Python libraries can provide you with pre-written function packages instantly. This saves a lot of time and improves productivity. Each library addresses a specific domain, e.g. machine learning, data science, data visualization, image and data manipulation applications, ...

To use a library, simply **import** it into your code.

Find in the cell below a very basic example of how a specific library can help save time and energy.

Let's try to compute the sum of an array

In basic Python we would use the following code

```python
def compute_mean(my_array):
    """
    This method counts the number of elements in an array
    Then computes the sum of the elements using the built-in method "sum"
    Finally, it outputs the mean value by dividing by the number of elements
    """
    nb_elements = len(my_array)  # count number of elements
    print("The array contains {0} elements".format(nb_elements))
    mean_value = (
        sum(my_array) / nb_elements
    )  # compute the sum, then divide by the number of elements
    return mean_value
```

```python
A = [1, 2, 3]
print(A)
print("The mean of the array is:", compute_mean(A))
```

However, the library "Numpy" offers many built-in methods to help

```python
import numpy

# Create the same array with Numpy
A = numpy.array([1, 2, 3])
mean_val = numpy.mean(A)
print("The mean is: {0}".format(mean_val))
```

# Work with Earth observation products

This second part of the demo explains how Python can be used to compute a vegation index on Sentinel-2 products

NDVI measures the ratio of the reflective difference in the red and near-infrared portions of the spectrum to the sum of red and near-infrared reflectance. Green, healthy vegetation reflects light in the near-infrared portion of the spectrum and absorbs red light. NDVI ranges from values of 1.0 to -1.0 where larger, positive values indicate green vegetation.

NDVI is used in many different applications. Specifically, it is an important indicator of the amount and quality of forage for wildlife and livestock. NDVI is also used to assess agricultural fields and identify the location and extent of surface water.

## Import & open products

Let's open a Sentinel-2 L1C product somewhere in the north of **CAMEROON** (Lake Lagdo) with [rasterio](https://rasterio.readthedocs.io/en/latest/) and Xarray.

We will fetch only red and nir bands (two separate files) to later compute the vegetation index using Normalized Difference Vegetation Index (NDVI)

$$\text{NDVI} = \frac{NIR - R}{NIR + R}$$

Where NIR denotes the Near Infra Red band and R, the Red Band

```python
import os
import rasterio
from rasterio.plot import show
import folium
from ipyleaflet import Map, basemaps, LayersControl
import xarray as xr

path = "./Products/Cameroon/"
file_red = "B04_crop_epsg4326.tif"  # Name of red band file
file_nir = "B08_crop_epsg4326.tif"  # Name of NIR band file
product_red = os.path.join(path, file_red)  # Create full path to file
product_nir = os.path.join(path, file_nir)  # Create full path to file
```

## Use Xarray to open and manage images

Xarray is a Python package that allows working with labelled multi-dimensional arrays.

Xarray introduces labels in the form of dimensions, coordinates and attributes on top of raw NumPy-like arrays, which allows for a more intuitive, more concise, and less error-prone developer experience. The package includes a large and growing library of domain-agnostic functions for advanced analytics and visualization with these data structures.

Let's open the red and nir products using Xarray

The products are 10980x10980 pixels, single band images. So, we'll import 1 band only, and select x & y dimensions to have 4*4 chunks of data

```python
import rioxarray as rxr

del img_crop  # clear memory
red = rxr.open_rasterio(product_red, chunks={"band": 1, "y": 250, "x": 250})
nir = rxr.open_rasterio(product_nir, chunks={"band": 1, "y": 250, "x": 250})
```

A Xarray is represented as shown below

```python
nir
```

## Compute NDVI

The NDVI can be computed from 2 separate Xarray DataArrays, by applying directly mathematical operations on the Xarrays

```python
NDVI = (nir - red) / (nir + red)
```

Note that the computed NDVI is the same type as the objects used : an Xarray with the same dimensions

```python
NDVI
```

However, Xarrays can be concatenated in a single structure.

We will concatenate nir and red bands into a single Xarray

```python
nir["band"] = ["nir"]  # Set band name to "nir" instead of int "1"
red["band"] = ["red"]  # Set band name to "red" instead of int "1"
data_xarray = xr.concat([nir, red], dim="band")
```

The dimensions of the Xarray are now 2 x 1000 x 1000 (2 bands)

```python
data_xarray
```

Please notice in the representation above that the structure is still the same size along x and y, but it now has 2 dimensions along the "band" dimension !

The NDVI can be computed as before but working with 2 dimensions of the same Xarray

```python
# Recall that : NDVI = (nir - red) / (nir + red)
NDVI_xarray = (data_xarray.sel(band="nir") - data_xarray.sel(band="red")) / (
    data_xarray.sel(band="nir") + data_xarray.sel(band="red")
)
NDVI_xarray
```

Display both computed NDVI side by side using matplotlib

This  areas of healthy vegetation with high NDVI values will appear greener in the image below.

```python
NDVI_small = NDVI.fillna(0)
NDVI_small_b = NDVI_xarray.fillna(0)
```

```python
from matplotlib import pyplot as plt

fig, axes = plt.subplots(ncols=2, sharey=True)
fig.suptitle("Both methods give the same NDVI result")
im = NDVI_small.plot(ax=axes[0], add_colorbar=False, cmap="Greens")
axes[0].set_title("Two Xarrays")
axes[0].yaxis.set_visible(False)
axes[0].xaxis.set_visible(False)

im2 = NDVI_small_b.plot(ax=axes[1], add_colorbar=False, cmap="Greens")
axes[1].set_title("Single Xarray")
axes[1].yaxis.set_visible(False)
axes[1].xaxis.set_visible(False)

plt.tight_layout()
```

## Compare performance of both methods

First method

```python
%%timeit -r 2 -n 100
(nir - red) / (nir + red)
```

```python
%%timeit -r 2 -n 100
(data_xarray.sel(band="nir") - data_xarray.sel(band="red")) / (data_xarray.sel(band="nir") - data_xarray.sel(band="red"))
```

Note that the execution time is faster when using multiple dimensions of the same Xarray
